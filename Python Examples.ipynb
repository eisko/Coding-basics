# Python Examples
This notebook contains examples of how to load, process, and visualize data in python.

## 1. Load packages

```python
import pandas as pd # The pandas packages allows easier visualization and processing of dataframes/csv files
# pandas offers similar functionality to tidyverse packages in R

import numpy as np # used to create arrays (i.e. matrices) of various sizes
# package contains functions that can be applied to pandas dataframes

import matplotlib.pyplot as plt # offers basic plotting capabilities in python

import seaborn as sns # another plotting package, slightly different to matplotlib
# offers more 'attractive' presets and handles data and plotting more similar to ggplot package in R

%matplotlib inline 
# this ensures graps can be visualized in notebook when you run code chunks
```

## 2. Import data and explore data

```python
# load csv or other files (e.g. tsv) can be done with this function
# will load file into a pandas dataframe object
# 
data = pd.read_csv('Pos0genesbycellID.txt', sep=',')
data
```

## 3. Graph Data
Can use either matplotlib or seaborn, here are examples of plots using matplotlib

### Scatterplot

```python
# scatterplot using matplotlib

plt.scatter(data['cellID'], data['num_rolonies'])
plt.title('Scatter Plot')
plt.ylabel('y value')
```

```python
# scatterplot using seaborn

sns.scatterplot(data=data, x="cellID", y="num_rolonies", hue="gene")
```

### Histogram

```python
# matplotlib
plt.hist(data['num_rolonies'],bins=140)
# bins does not need to be specified, but found that figure looked weird without it
```

```python
# plot x axis on log scale
sns.displot(data['num_rolonies'], bins=140)
plt.title("Histogram of Rolony Numbers")

# to save figure:
plt.savefig('rolony_histogram.jpg', dpi=300, bbox_inches='tight')
```
